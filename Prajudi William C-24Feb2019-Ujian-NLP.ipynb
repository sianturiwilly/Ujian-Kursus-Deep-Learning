# Ujian 1
Ujian ini menggunakan data dalam bentuk bahasa inggris, yaitu menggunakan data:

https://edition.cnn.com/2019/02/23/americas/venezuela-brazil-violence-intl/index.html

### 1. Crawling
-- tuliskan penjelasan kalian

```python
## kode untuk crawling
# library yang dibutuhkan
import feedparser
import os
import time
from boilerpipe.extract import Extractor


#english
url_eng="https://edition.cnn.com/2019/02/23/americas/venezuela-brazil-violence-intl/index.html"
extractor_eng = Extractor(extractor='ArticleExtractor', url=url_eng)
content_eng = extractor_eng.getText() 
content_eng=content_eng.lower()
print(content_eng)
```

### 2. Tokenization
-- tuliskan penjelasan kalian

```python
## kode tokenization
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize

## build to tokenize (raw)
array_word_eng=word_tokenize(content_eng)
print(array_word_eng)

## build to tokenize with sentence 
array_kalimat=sent_tokenize(content_eng)
print(array_kalimat)
```

### 3. Analisis Frequensi Distribusi
-- tuliskan penjelasan kalian

```python
## kode frekuensi distribusi

# menghitung berapa jumlah distribusi per kata
import nltk
counts  = nltk.FreqDist(content_eng)
counts
```

### 4. Analisis Word Cloud sebelum stopword removal
-- tuliskan penjelasan kalian

```python
## analisis word cloud sebelum stopword removal

%matplotlib inline
from pprint import pprint
import matplotlib.pyplot as plt
from wordcloud import WordCloud



cloud_eng = WordCloud(max_words=1200)
cloud_eng.generate_from_text(content_eng)
plt.figure(figsize=(12,8))
plt.imshow(cloud_eng)
plt.axis('off')
plt.show()
```

### 5. Implementasi Stopword removal
--tuliskan penjelasan kalian

```python
## kode
#indonesia
import string

f=open("data/StopWordEN")
stopword_eng=f.read()
eng_stopwords  =word_tokenize(stopword_eng)
eng_stopwords
punctuation=string.punctuation
additional_punctuation=[".","-","``"]

array_punctuation=word_tokenize(punctuation)
learn_word=[] #this can be learn hehe improve
stopword_eng=eng_stopwords+array_punctuation +additional_punctuation+learn_word
print(stopword_eng)
```

### 6. Analisis Word Cloud setelah stopword removal

```python
## analisis word cloud setelah stop word

filter_word=[word for word in array_word_eng if word not in stopword_eng]
filter_word

%matplotlib inline
from pprint import pprint
import matplotlib.pyplot as plt
from wordcloud import WordCloud

after_stop = ' '.join(filter_word)

cloud = WordCloud(max_words=1200)
cloud.generate_from_text(after_stop)
plt.figure(figsize=(12,8))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```
